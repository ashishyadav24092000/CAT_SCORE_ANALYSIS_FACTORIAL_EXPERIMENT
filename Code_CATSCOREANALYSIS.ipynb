```python
import pandas as pd
```

```python
path = '/home/ashish/projects/CATSCORE_ANALYSIS_Factorial_design_2wayanova/Twoway.xlsx'
data_frame = pd.read_excel(path)
data_frame
```

```python
import scipy
from statsmodels.formula.api import ols
import statsmodels.api as sm
```

```python
form_show = 'Value ~ C(prep_pro) +  C(college) + C(prep_pro):C(college)'
model = ols(form_show,data_frame).fit()
model.summary()
```

```python
anova_table = sm.stats.anova_lm(model,typ=2)
anova_table
```

# CONCLUSION::
## Factor A = preparation program type(3 treatments).
## Factor B = college background type(3 treatments).

        Since,
            1. p-value(factor A)=0.299 > alpha=0.05 ==> accept null hypothesis.
            2. p-value(Interaction factor)=0.350 > alpha=0.05 ==>accept null hypothesis.
            3.p-value(Factor B)=0.005 < alpha=0.05 ==>reject null hypothesis(B-factor).

# We can confidently say with a significance level of 5% that only the factor B(the type of under graduate college) can significantly effect  the CAT Score.
